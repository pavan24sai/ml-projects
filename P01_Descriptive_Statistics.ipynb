```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from ggplot import mtcars
```

```python
mtcars.index = mtcars["name"]
# Also remove the name column from mtcars DataFrame since it's not required for futher data analysis
mtcars = mtcars.drop(columns=["name"], axis=1)
```

```python
mtcars.head()
```

```python
# see all the functions of mtcars
#dir(mtcars)
```

```python
# Calculate the mean of each of the columns of mtcars.
mtcars_means = mtcars.mean(numeric_only=True)
mtcars_means
```

```python
# Get the mean of each row (car) and store it in variable 'mtcarsMeanCarWise'.
mtcarsMeanCarWise = mtcars.mean(axis=1)
mtcarsMeanCarWise
```

```python
# Get the median of each column and store it in variable 'medianCarsFeatures'.
medianCarsFeatures = mtcars.median(axis=0)
medianCarsFeatures
```

```python
# Plot the histograms for each of the columns in mtcars 
# using hist() function with bins as 10 and figsize of (20,15).
mtcars.hist(bins=10, figsize=(20, 15))
plt.show()
```

```python
# Plot the density curves for each column of mtcars using 'for' loop and plot command. 
# You can plot density by calling plot function with arguments kind="density" and figsize=(2,2).
for c in mtcars.columns:
    mtcars[c].plot(kind="density", figsize=(2,2))
    plt.title(c)
    plt.show()
```

```python
# Plot the correlation for the following attributes: mpg, disp, hp, wt, qsec. 
# You can use the scatter_matrix command present in pandas.plotting library
columns_to_correlate = ['mpg', 'disp', 'hp', 'wt', 'qsec']
pd.plotting.scatter_matrix(mtcars[columns_to_correlate])
```

```python
# Calculate the skeweness of each column of 'mtcars'
# We use skewness as a measure of symmetry. 
# -> If the skewness of is zero then the distribution represented is perfectly symmetric. 
# -> If the skewness is negative, then the distribution is skewed to the left, while 
# -> if the skew is positive then the distribution is skewed to the right. 
# 
# Hence, it can also be considered as a measure of normality since normal distributions are symmetric and hence have a zero skew.
mtcars.skew()
```

```python
# Use describe and quantile to calculate the Inter-Quantile spread of 'mpg' attribute of mtcars.
# Method 1: Using describe()
stats = mtcars['mpg'].describe()
q1 = stats['25%']
q3 = stats['75%']
iqr_describe = q3 - q1

# Method 2: Using quantile() directly
q1_direct = mtcars['mpg'].quantile(0.25)
q3_direct = mtcars['mpg'].quantile(0.75)
iqr_quantile = q3_direct - q1_direct

print(f"IQR using describe: {iqr_describe}")
print(f"IQR using quantile: {iqr_quantile}")
```

```python
# Plot the box plot for 'mpg' column of mtcars.
mtcars['mpg'].plot(kind="box")
```

```python
# Calculate the mean, standard deviation and variance of the 'mpg' column
mpg_mean  = mtcars['mpg'].mean()
mpg_stdev = mtcars['mpg'].std()
mpg_var   = mtcars['mpg'].var()

print(f"Mean of 'mpg' column: {mpg_mean}")
print(f"Standard Deviation of 'mpg' column: {mpg_stdev}")
print(f"Variance of 'mpg' column: {mpg_var}")
```

```python
# Extract cars with automatic transmission and cars with manual transmission into separate variables 
# 'mtCarsAutomatic' and 'mtCarsManual' respectively
mtCarsAutomatic = mtcars[mtcars['am'] == 0]
mtCarsManual    = mtcars[mtcars['am'] == 1]
```

```python
# Plot the box plot for 'mpg' (miles per gallons) for each of the subsets of cars grouped by transmission (automatic/ manual)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
axes = axes.flatten()
mtCarsAutomatic['mpg'].plot(kind="box", ax=axes[0])
axes[0].set_title('mpg - Automatic')
mtCarsManual['mpg'].plot(kind="box", ax=axes[1])
axes[1].set_title('mpg - Manual')
```

```python
# Can you come up with similar box plots to observe how other variables might be significantly impacting Miles/(US) gallon? 
# What are the variables we should track if we need to predict mpg (Miles/ (US) gallon) ?
# plot for different gears - 3, 4, 5
mtCars_Gears3 = mtcars[mtcars['gear'] == 3]
mtCars_Gears4 = mtcars[mtcars['gear'] == 4]
mtCars_Gears5 = mtcars[mtcars['gear'] == 5]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 3))
axes = axes.flatten()
mtCars_Gears3['mpg'].plot(kind="box", ax=axes[0])
axes[0].set_title('mpg - Number of Gears = 3')
mtCars_Gears4['mpg'].plot(kind="box", ax=axes[1])
axes[1].set_title('mpg - Number of Gears = 4')
mtCars_Gears5['mpg'].plot(kind="box", ax=axes[2])
axes[2].set_title('mpg - Number of Gears = 5')
```

```python
# Generate 10000 random normally distributed numbers and plot the density curve along with the median and mean.
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Generate normal distribution data
normal_data = np.random.normal(size=10000)
normal_mean = np.mean(normal_data)
normal_median = np.median(normal_data)

# Plot normal distribution
pd.Series(normal_data).plot(kind='density', ax=ax1)
ax1.axvline(normal_mean, color='red', linestyle='--', label=f'Mean: {normal_mean:.4f}')
ax1.axvline(normal_median, color='green', linestyle='-.', label=f'Median: {normal_median:.4f}')
ax1.set_title('Normal Distribution')
ax1.legend()

# Generate exponential distribution data
exponential_data = np.random.exponential(size=10000)
exponential_mean = np.mean(exponential_data)
exponential_median = np.median(exponential_data)

# Plot exponential distribution
pd.Series(exponential_data).plot(kind='density', ax=ax2)
ax2.axvline(exponential_mean, color='red', linestyle='--', label=f'Mean: {exponential_mean:.4f}')
ax2.axvline(exponential_median, color='green', linestyle='-.', label=f'Median: {exponential_median:.4f}')
ax2.set_title('Exponential Distribution')
ax2.legend()

plt.tight_layout()
plt.show()
```
